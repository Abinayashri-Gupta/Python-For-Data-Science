```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn-whitegrid')
```

```python
x = [5,11,8,5,2,3,6,7,8,10]
y = [3,4,5,16,18,9,2,3,4,32]
plt.scatter(x,y,c='magenta')
```

```python
x = [5,11,8,5,2,3,6,7,8,10]
y = [3,4,5,16,18,9,2,3,4,32]
plt.scatter(x,y,c='crimson')
plt.xlabel('data-x')
plt.ylabel('data-y')
plt.title('Scatter Plot')
plt.show()
```

```python
x = np.linspace(0,10,30)
y = np.sin(x)

print(plt.plot(x,y,'*',c='purple'))
print(plt.plot(x,y,c='purple'))
```

```python
print(plt.plot(x,y,'^',c='purple'))
```

```python
print(plt.plot(x,y,'d',c='purple'))
```

```python
print(plt.plot(x,y,'s',c='purple'))
```

```python
import numpy as np
from matplotlib import pyplot as plt

theta = np.linspace(0, 2*np.pi, 1000)

r = 5 - 5 * np.sin(theta)

plt.polar(theta, r, 'r')

plt.show()
```
