```python
from mpl_toolkits import mplot3d
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()
ax = plt.axes(projection='3d')
```

```python
#Three-Dimensional Points and Lines:
```

```python
ax = plt.axes(projection='3d')
# Data for a three-dimensional line
zline = np.linspace(0, 19, 2000)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline, yline, zline, 'purple')
```

```python
#Three-dimensional Countour
```

```python
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)#change the shape and view
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x');
ax.set_ylabel('y');
ax.set_zlabel('z');
```

```python
ax.view_init(80,55)
fig#change the view of the image
```

```python
#Wireframes and Surface Plots
```

```python
fig = plt.figure()
ax = plt.axes(projection = '3d')
ax.plot_wireframe(X, Y, Z, color='purple')
ax.set_title('wireframe');
```

```python
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
 cmap='viridis', edgecolor='purple')
ax.set_title('surface');
```
