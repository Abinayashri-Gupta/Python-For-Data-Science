# Creating a DataFrame Using Pandas

```python
import pandas as pd
```

```python
df = pd.DataFrame({
'Merchandise': ['Shirt', 'Lightstick', None, 'Shoes', None, 'Tank Tops', 'Pants', 'Hair Bands', 'Bracelets'],
'Team Name': ['CSK', 'LSG', 'RCB', 'GT', 'MI', None, 'RR', 'ASP', 'CFC'],
'Price': [1200, 3500, None, 1200, None, 250, 7000, 12000, 799]
})
```

```python
df
```

### isnull()

```python
df.isnull()
```

### notnull()

```python
df.notnull()
```

### dropna()

```python
df.dropna()
```

### replace(),to_replace,value

```python
df.replace(to_replace=1200,value=1800)
```
